```python
import requests
import pandas as pd
import re
from string import whitespace
from bs4 import BeautifulSoup
```

```python
url = 'https://www.goszakup.gov.kz/ru/registry/rqc'
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
comps = soup.find('table', class_="table table-bordered table-hover")

headers = []
for i in comps.find_all('th'):
    title = i.text
    headers.append(title)

mydata = pd.DataFrame(columns = headers)
for j in comps.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(mydata)
    mydata.loc[length] = row
mydata1 = mydata[['Наименование потенциального поставщика','БИН/ИИН']]
mydata2=mydata1.replace('\n\n','',regex=True)
mydata2
```

```python
mydata2.to_csv('Adata_companies.csv', encoding='utf-16')
```
